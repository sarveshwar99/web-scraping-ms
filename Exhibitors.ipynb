```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html=urlopen("https://www.lineapelle-fair.it/tools/catalogo-lineapelle/espositore.php?iq=123zz5&ediz=95")
```

```python
soup=BeautifulSoup(html,'html5lib')
```

```python
print(soup.prettify())
```

```python
soup.find('div',id="tabs-2")
```

```python
def f_title(soup):
    return(soup.find('div',id="tabs-2").h1.get_text())
```

```python
f_title(soup)
```

```python
def f_email_address(soup):
    return(soup.find('div',id="tabs-2").find('a',{'href':re.compile('.@.')})['href'])
```

```python
f_email_address(soup)
```

```python
def f_website(soup):
    return(soup.find('div',id="tabs-2").find('a',{'href':re.compile('http.')})['href'])
```

```python
f_website(soup)
```

```python
html2=urlopen('https://www.lineapelle-fair.it/tools/catalogo-lineapelle/?l=en&t=1')
```

```python
soup2=BeautifulSoup(html2,'html5lib')
```

```python
print(soup2.prettify())
```

```python
page_1=[line['href'] for line in soup2.table.findAll('a')]
```

```python
len(page_1)
```

```python
titles=[]
emails=[]
websites=[]
```

```python
for line in page_1:
    url=urlopen('https://www.lineapelle-fair.it/tools/catalogo-lineapelle/'+line)
    soup=BeautifulSoup(url,'html5lib')
    titles.append(f_title(soup))
    emails.append(f_email_address(soup))
    websites.append(f_website(soup))
```
