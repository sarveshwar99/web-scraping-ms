```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
```

```python
headers = {'user-agent': 'my-app/0.0.2'}
```

```python
html=requests.get('https://www.amazon.com/Card-Catalog-Notecards-Library-Congress/dp/1452114323/ref=sr_1_19?s=books&ie=UTF8&qid=1538052883&sr=1-19&keywords=book+catalog',headers=headers)
```

```python
print(soup.prettify())
```

```python
def f_isbn_10(soup):
    x=soup.find('div',{'class':'content'}).find('b',text='ISBN-10:')
    if x:
        return re.sub('ISBN-10: ','',x.parent.get_text())
    else:
        return None
```

```python
f_isbn_10(soup)
```

```python
def f_isbn_13(soup):
    x=soup.find('div',{'class':'content'}).find('b',text='ISBN-13:')
    if x:
        return re.sub('ISBN-13: ','',x.parent.get_text())
    else:
        return None
```

```python
f_isbn_13(soup)
```

```python
def f_title(soup):
    x=soup.find('span',{'id':['productTitle','title']})
    if x:
        return x.get_text()
    else:
        return None
```

```python
f_title(soup)
```

```python
def f_dimensions(soup):
    x=soup.find('b',text=re.compile('\s+Product Dimensions:\s+'))
    if x:    
        return re.sub('ProductDimensions:','',re.sub(r"[\n\t\s]*", "", x.parent.get_text()))
    else:
        return None
```

```python
f_dimensions(soup)
```

```python
def f_shipping_weight(soup):
    x=soup.find('b',text=re.compile('Shipping Weight:'))
    if x:    
        return re.sub('Shipping Weight: ','',x.parent.get_text())
    else:
        return None
```

```python
f_shipping_weight(soup)
```

```python
def f_feedback_rating(soup):
    x=soup.find('i',{'class':'a-icon a-icon-star a-star-4-5'})
    if x:    
        return x.find('span',{'class':'a-icon-alt'}).get_text()
    else:
        return None
```

```python
f_feedback_rating(soup)
```

```python
def f_sales_rank(soup):
    try:
        x=soup.find('li',{'id':'SalesRank'})
        if x:
            return re.sub('\n|\(','',soup.find('li',{'id':'SalesRank'}).a.previous_element+soup.find('li',{'id':'SalesRank'}).a.next_element)+(re.sub('\n|\xa0',' ',soup.find('li',{'id':'SalesRank'}).ul.get_text())).strip()
        else:
            return None
    except AttributeError:
        return re.sub('\n|\(','',soup.find('li',{'id':'SalesRank'}).a.previous_element+soup.find('li',{'id':'SalesRank'}).a.next_element)
    
```

```python
f_sales_rank(soup)
```

```python
def f_language(soup):
    x=soup.find('b',text=re.compile('Language:'))
    if x:    
        return re.sub('Language: ','',x.parent.get_text())
    else:
        return None
```

```python
f_language(soup)
```

```python
def f_hardcover_paperback(soup):
    x=soup.find('b',text=re.compile('Hardcover:|Paperback:'))
    if x:    
        return x.parent.get_text()
    else:
        return None
```

```python
f_hardcover_paperback(soup)
```

```python
Title=[]
ISBN_10=[]
ISBN_13=[]
Dimensions=[]
Shipping_Weight=[]
Feedback_Rating=[]
Sales_Rank=[]
Language=[]
Hardcover_Paperback_Pages=[]
```

```python
def df_f(links):
    for link in links:
        time.sleep(5)
        html=requests.get(link,headers=headers)
        soup=BeautifulSoup(html.text,'lxml')
        
        Title.append(f_title(soup))
        ISBN_10.append(f_isbn_10(soup))
        ISBN_13.append(f_isbn_13(soup))
        Dimensions.append(f_dimensions(soup))
        Shipping_Weight.append(f_shipping_weight(soup))
        Feedback_Rating.append(f_feedback_rating(soup))
        Sales_Rank.append(f_sales_rank(soup))
        Language.append(f_language(soup))
        Hardcover_Paperback_Pages.append(f_hardcover_paperback(soup))
    
```
