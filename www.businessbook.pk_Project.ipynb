```python
categ_links_list=[]
def f_category_links(links):
    for link in links:
        html=requests.get(link,headers=headers,proxies=proxies)
        soup=BeautifulSoup(html.text,'lxml')
        categ_links_list.append(f_links(soup2))
        time.sleep(4)
```

```python
exporters_links_list=[i for b in map(lambda x:[x] if not isinstance(x, list) else x, categ_links_list) for i in b]
```

```python
url3=exporters_links_list[0]
```

```python
html3=requests.get(url3,headers=headers,proxies=proxies)
soup3=BeautifulSoup(html3.text,'lxml')
```

```python
html3.request.headers
```

```python
def f_company_name(soup):
    return soup.find('img',{'class':'img-thumbnail'})['title']
```

```python
f_company_name(soup3)
```

```python
def f_mobile_phone(soup):
    return soup.find('li',{'class':'list-group-item phone-icons'}).text if soup.find('li',{'class':'list-group-item phone-icons'}) else None
```

```python
f_mobile_phone(soup3)
```

```python
def f_website(soup):
    return soup.find('a',{'itemprop':'url'},rel='nofollow').text if soup.find('a',{'itemprop':'url'},rel='nofollow') else None
```

```python
f_website(soup3)
```

```python
def f_city(soup):
    return soup.find('span',{'itemprop':'addressLocality'}).text
```

```python
f_city(soup3)
```

```python
def f_category(soup):
    return soup.find('span',{'itemprop':'name'}).text
```

```python
f_category(soup3)
```

```python
Company_Name=[]
Mobile=[]
Phone=[]
Website_URL=[]
City=[]
Category=[]
```

```python
html=requests.get('https://www.businessbook.pk/',headers=headers,proxies=proxies)
time.sleep(3)
html=requests.get('https://www.businessbook.pk/category/exporters-in-pakistan-1263',headers=headers,proxies=proxies)
time.sleep(3)
for link in exporters_links_list:
    html=requests.get(link,headers=headers,proxies=proxies)
    soup=BeautifulSoup(html.text,'lxml')
    Company_Name.append(f_company_name(soup))
    Mobile.append(f_mobile_phone(soup))
    #Phone.append(f_company_name(soup))
    Website_URL.append(f_website(soup))
    City.append(f_city(soup))
    Category.append(f_category(soup))
    time.sleep(2)
    
```
