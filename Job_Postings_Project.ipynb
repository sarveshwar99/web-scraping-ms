```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import re
import pandas as pd
```

```python
html=requests.get(url)
```

```python
soup=BeautifulSoup(html.text,'lxml')
```

```python
last_page=int(soup.findAll('li',{'class':'page'})[-1].text.strip())
```

```python
print(soup.prettify())
```

```python
pages=np.arange(2, last_page+1)
```

```python
l_pages=list(pages)
l_pages
```

```python
print('https://yellow.local.ch/en/q/Vaud%20(Canton)?cat=1499&page='+str(l_pages[0])+'&rid=88-Y&typeref=bus&what=')
```

```python
links=['https://yellow.local.ch/en/q/Vaud%20(Canton)?cat=1499&page='+str(link)+'&rid=88-Y&typeref=bus&what=' for link in l_pages]
```

```python
links.insert(0,url)
```

```python
[print(link) for link in links]
```

```python
headers={'User-Agent':'micky gall 2565'}
```

```python
proxies = {
  "http": "http://user:pass@10.10.1.10:3128/"
}
```

```python
job_links=[]
for link in links:
    html=requests.get(link,headers=headers)
    soup=BeautifulSoup(html.text,'lxml')
    job_links.append([line['href'] for line in soup.findAll('a',{'class':'listing-link clearfix'})])
```

```python
len(job_links)
```

```python
job_links=[i for b in map(lambda x:[x] if not isinstance(x, list) else x, job_links) for i in b]
```

```python
html2=requests.get(job_links[0],headers=headers)
```

```python
soup2=BeautifulSoup(html2.text,'lxml')
```

```python
soup2.find('a',{'class':'detail-entry-title'})
```

```python
def f_company_name(soup):
    return soup.find('span',{'class':'detail-entry-title'}).text.strip()
```

```python
f_company_name(soup2)
```

```python
def f_website(soup):
    return soup.find('a',{'class':'js-kpi-url redirect'}).text.strip() if soup.find('a',{'class':'js-kpi-url redirect'}) else None
```

```python
f_website(soup2)
```

```python
def f_decision_maker(soup):
    if soup.find('div',{'class':'lui-card-body'}).find('div',{'class':'row vcard h-card'}).find('p',{'class':'lead lui-margin-bottom-xs name-additional'}):
        return soup.find('div',{'class':'lui-card-body'}).find('div',{'class':'row vcard h-card'}).find('p',{'class':'lead lui-margin-bottom-xs name-additional'}).text.strip()
    else:
        return None
```

```python
print(soup.prettify())
```

```python
soup.find('div',{'class':'lui-card-body'}).find('div',{'class':'row vcard h-card'}).find('p',{'class':'lead lui-margin-bottom-xs name-additional'})
```

```python
f_decision_maker(soup)
```

```python
f_decision_maker(soup2)
```

```python
def f_title(soup):
    return re.findall('([a-zA-Z\s]+)\•{1}',soup2.find('div',{'class':'categories detail-header-categories hotellerie-suisse-hidden'}).text.strip() )[0].strip() if soup2.find('div',{'class':'categories detail-header-categories hotellerie-suisse-hidden'}) else None
```

```python
f_title(soup2)
```

```python
def f_email(soup):
    return soup.find('div',{'class':'lui-margin-vertical-zero contact-value-container lui-text-overflow'}).find('a',class_="js-kpi-email").text.strip() if soup.find('div',{'class':'lui-margin-vertical-zero contact-value-container lui-text-overflow'}).find('a',class_="js-kpi-email") else None
```

```python
f_email(soup)
```

```python
def f_phone(soup):
    return soup.find('div',{'class':'lui-margin-vertical-zero contact-value-container'}).text.strip() if soup.find('div',{'class':'lui-margin-vertical-zero contact-value-container'}) else None
```

```python
f_phone(soup2)
```

```python
Company_Name=[]
Website=[]
Decision_maker=[]
Job_title=[]
Email_address=[]
Phone=[]
```

```python
for link in job_links[:100]:
    html=requests.get(link,headers=headers,proxies=proxies)
    soup=BeautifulSoup(html.text,'lxml')
    Company_Name.append(f_company_name(soup))
    Website.append(f_website(soup))
    Decision_maker.append(f_decision_maker(soup))
    Job_title.append(f_title(soup))
    Email_address.append(f_email(soup))
    Phone.append(f_phone(soup))
```

```python
html=requests.get(job_links[1],headers=headers)
soup=BeautifulSoup(html.text,'lxml')
```

```python
print(soup.prettify())
```

```python
df=pd.DataFrame({'Company_Name':Company_Name,'Website':Website,'Decision_maker':Decision_maker,'Job_title':Job_title,'Email_address':Email_address,'Phone':Phone})
```
